```python
#Zscore

plt.boxplot(df['Writing_score'])
```

```python
ul=df['Writing_score'].mean()+3*df['Writing_score'].std()

print(ul)

ll=df['Writing_score'].mean()-3*df['Writing_score'].std()
print(ll)
```

```python
#tripping

#df1=df.loc((df['Writing_score']<=ul) & (df['Writing_score']>=ll))
#df1

df1=df.loc[(df['Writing_score']<=ul) & (df['Writing_score']>=ll)]
df1
```

```python
print("before triming",len(df))
print("After Triming",len(df1))
```

```python
print("number of outliers",len(df)-len(df1))
```
